```python
import pandas as pd
```

```python
# Create a dataset with 200 rows for datetime practice
df = pd.DataFrame({
    'date_created': pd.date_range('2023-01-01', periods=200, freq='D'),
    'date_modified': pd.date_range('2023-06-20', periods=200, freq='12h'),
    'value': range(1, 201),
    'category': ['A', 'B', 'C', 'D'] * 50
})

print(df.info())
df.head()
```

```python
# convert datetime into strings
df['date_created'].dt.strftime('%d-%m-%Y')
```

```python
dt_accessor = df["date_created"].dt
dt_accessor.year
dt_accessor.month
dt_accessor.day
dt_accessor.day_of_week >= 5
```

```python
# extract date and time components
dt_accessor.time
dt_accessor.date
```

```python
dt_accessor.quarter.astype("str") + "Q " + dt_accessor.year.astype("str")
```

```python
df["date_created"] + pd.Timedelta("7 days") + pd.Timedelta("2 hours") + pd.Timedelta("30 minutes") + pd.Timedelta("15 seconds")
```

```python
pd.date_range(start="2024-01-01", end="2024-12-31", freq="h")

# month-end anchored (12 months)
pd.date_range(start="2024-01-31", periods=12, freq="ME")
```

```python
# Add a datetime string column and inject missing values
df["datetime_miss"] = df["date_modified"].dt.strftime("%Y-%m-%d %H:%M:%S")
df.loc[df.index % 7 == 0, "datetime_miss"] = pd.NA
df
```

```python
df["datetime_miss"] = pd.to_datetime(df["datetime_miss"])
df
```

```python
df["datetime_miss"].bfill() # bfill fills with the last observed value (backward fill)
df["datetime_miss"].ffill() # ffill fills with the next observed value (forward fill)

# limit to at most N consecutive gaps filled
df["datetime_miss"].ffill(limit=3)

# combined: ffill first, then bfill for any leading NaN
df["datetime_miss"].ffill().bfill()
```

```python
pd.Timestamp("2024-01-01 00:00:00")
pd.to_datetime("2024-01-01 00:00:00")
```

```python
# filling missing values with a specific value
df["datetime_miss"].fillna(pd.to_datetime("2024-01-01 00:00:00"))

# with the most repeated value
df["datetime_miss"].fillna(df["datetime_miss"].mode()[0])
```

```python
# fillna with interpolation
df["datetime_miss"].interpolate(method="linear")
```

```python
df["date_shifted"] = df["date_created"].shift(2) # shift the values down by 2 rows
df["date_diff"] = df["date_created"].diff(1) # difference between the current and previous row
df
```

```python
# rolling windows
df.value.rolling(window=7).mean() # 7-day rolling mean
df.value.rolling(window=30).sum() # 30-day rolling sum
```

```python
df["value"].expanding().sum() # expanding is similar to rolling, but it uses all previous values instead of a fixed window size
```

```python
# query
df[df["date_created"].dt.month == 1] # all rows where the month is January
df.query('date_created.dt.month == 1') # same as above, but using query method

df[(df["date_created"] > "2023-01-01") & (df["date_created"] < "2023-02-01")] # all rows where the date is in January 2023
df[df["date_created"].between("2023-01-01", "2023-02-01")] # same as above, but using between method
df.query('date_created.between("2023-01-01", "2023-02-01")') # same as above, but using query method
```

```python
# use where, the difference is that where returns the original value if the condition is True, and NaN otherwise,
#  while query returns only the rows that satisfy the condition
df["date_created"].where(df["date_created"].dt.month == 1) # all rows where the month is January, NaN otherwise
```
